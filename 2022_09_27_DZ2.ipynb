```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as slv
```

## Часть 1

ФПД для уравнения  
$$
f(y, B)=y-y^{2}-B
$$


```python
dy_dt = y - y**2 - B
```

```python
Bs = np.linspace(-3,5,21) # задаёт точки расположения векторов горизонтль
ys = np.linspace(-3,3,21) # задаёт точки вертикаль


dydt_zeros =  np.zeros((21,21)) # задаёт абсциссы векторов 
dydt =  np.zeros((21,21))

for i,  B in enumerate(Bs):    
    for j, y in enumerate(ys):
        dydt[j, i] = y - y**2 - B # j первое потому в nd array - номер ряда (ордината)
        
        
plt.figure(figsize = (15, 10))
plt.quiver(Bs, ys, dydt_zeros, dydt, units='xy', scale=10, width = 0.015, headwidth = 5)

plt.xlabel('B')
plt.ylabel('y')
plt.show()
```

## Часть 2

Уравнение (3)
$$
\frac{d y}{d t}=y\left(d-k y^{2}\right)
$$


```python
Ds = np.linspace(-200,200,25) # задаёт точки расположения векторов горизонтль
ys  = np.linspace(-60, 60,25) # задаёт точки вертикаль
k = 0.05

dydt_zeros =  np.zeros((25, 25)) #  абсциссы векторов 
dydt =  np.zeros((25, 25)) #  ординаты векторов 

for i, d in enumerate(Ds):    
    for j, y in enumerate(ys):
        dydt[j, i] = y *(d - k*y**2) 
        
        
plt.figure(figsize = (15, 10))
plt.quiver(Ds, ys,  dydt_zeros, dydt, units='xy', headwidth = 5, color='#77d123')
plt.plot((-200, 200), (0, 0), label = 'y = 0')
plt.plot([k*y**2 for y in ys], ys, label= 'd =  k*y**2')
plt.scatter(0, 0,  marker='*', s=100, c = 'r') # сделать маркер точки бифуркации
plt.annotate('Точка бифуркации', (0, 0), xytext=(2, 2))
plt.annotate('Устойчивая ветвь \n Увеличение числа чёрных кошек', (50, 30), 
             xytext=(90, 30), arrowprops= {'width': 0.1, 'headwidth': 5}, 
             fontsize= 13)
plt.annotate('Устойчивая ветвь \n Увеличение числа белых кошек', (50, -30), 
             xytext=(70, -20), 
             arrowprops= {'width': 0.1, 'headwidth': 5}, 
             fontsize= 13)
plt.annotate('Устойчивая ветвь \n Преобладание серых кошек', (-30, 0), 
             xytext=(-150, 10), 
             arrowprops= {'width': 0.1, 'headwidth': 5}, 
             fontsize= 13)
plt.annotate('Неустойчивая ветвь \n Уменьшение числа серых кошек', (70, 0), 
             xytext=(90, 10),
             arrowprops= {'width': 0.1, 'headwidth': 5}, 
             fontsize= 13)

plt.xlabel('d')
plt.ylabel('y')
plt.legend()
plt.show()
```

Точка в бифуркации в d = 0  
При d < 0 существует только одна устойчивая точка, серый цвет шерсти  
При d > 0 появляются две устойчивые точки, с более чёргной и более белой шерстью. Точка "серой шерсти" становится неустойчивой.
