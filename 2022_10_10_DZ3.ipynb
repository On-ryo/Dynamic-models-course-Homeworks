## Часть 1  


****************************************************************************************  
### П.1

Отображение:

$$
x_{n+1}=r \cdot x_{n}\left(1-x_{n} / K\right)+x_{n}
$$


Неподвижные точки:
$$X^{*} = 0$$  
$$X^{*} = K$$

Производная:
$$F'(X)= -\frac{2 r x}{K}+r-1
$$

********************************************************************************  
Производная в неподвижных точках:

$$X^{*} = 0$$
$$F'(0) = r - 1 $$

Точка стабильна при  0 < r < 2 ???  
При x0 < K


$$X^{*} = K$$
$$F'(K) = {\infty}$$


Точка стабильна при  -2 < r <  0 ???  
И только с одной стороны от K (x0  < K)

****************************************************************************************   
### Пп. 2, 3


На графике ниже можно видеть, что при обоих вариантах K функция может сойтись либо к точке устойчивости x=0, либо к точке x=K

То, к какой из точек сойдётся (и сойдётся ли) функция, зависит от r и K  
Функция сходится к K при r в промежутке (0, 2),  любом (?) K  
Сходится к 0 при -2 < r < 0, x0 < K


```python
fu = lambda x,r,K: r*x *(1 - x/K) + x
rs = [-0.7, 0.7, 1, 1.5]
Ks = [5, 10]
x0 = 3
xns = [x0]

fig = plt.figure(figsize = (7, 5))
for r in rs:
    for K in Ks:
        for i in range(30):
                xns.append(fu(xns[-1], r, K))
        
        xn_1 = xns[1:]
        xn = xns[0:-1]
        plt.plot(xn_1, xn, label = f'r = {r}, K = {K}')
        xns = [x0]
        
        
xs = range(-1, 15, 1)
plt.plot(xs, xs)
plt.annotate('Точка  X = K = 10',
             (10, 10), xytext=(4, 13),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)
plt.annotate('Точка  X = K = 5',
             (5, 5), xytext=(-1, 8),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)
plt.annotate('Точка  X = 0',
             (0, 0), xytext=(2, 0),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)

plt.legend(bbox_to_anchor=(1, 1))
plt.xlabel('Xn')
plt.ylabel('Xn+1')
plt.grid(visible = True)
```

## Часть 2  
Уравнение взрыва численности насекомых

$$
\frac{d y}{d t}=r y\left(1-\frac{y}{k}\right)-\frac{y^{2}}{1+y^{2}}
$$

****************************************************************************************  
### П.1, 2

Стационарные ветви на фазопараметрической диаграмме:
$$y = 0 $$  

$$
r=\frac{y}{1+y^{2}-\frac{y}{k}-\frac{y^{3}}{k}}
$$


```python
K = 15
meshsize = 30
rs = np.linspace(0, 3, meshsize) # задаёт точки вертикаль
ys = np.linspace(0, 20, meshsize)
ys2 = np.linspace(0, 20, 300)
Rm, Ym  = np.meshgrid(rs, ys)

dy_dt_func = lambda r, y, K: r* y *(1 - y/K) - y**2 / (1 + y**2)

zeros = np.zeros((meshsize, meshsize))

plt.figure(figsize = (12, 8))
plt.quiver(Rm, Ym,  zeros, dy_dt_func(Rm, Ym, K), units='xy', scale = 6, color='#77d123') 

plt.plot([y/((1+y**2)*(1 -y/K)) for y in ys2], ys2, label = 'dy/dt = 0')
plt.plot(ys, np.linspace(0,0,30), label = 'y = 0')



plt.annotate('Точка бифуркации r  = 0.536 ',
             (0.536, 0.9), xytext=(0.8, 0.5),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)

plt.annotate('Точка бифуркации?',
             (0.27, 7.6), xytext=(0.5,8),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)

plt.annotate('Точка бифуркации r = 0, y = 0',
             (0, 0), xytext=(0.2, 3),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)


plt.title('Фазопараметрическая диаграмма, y от r')
plt.xlabel('r')
plt.ylabel('y')
plt.ylim(-1, 20)
plt.xlim(0, 1.5)
plt.legend()
plt.show()
```

### П. 5  

Кинетики y(t) при большом y и r проходящем через левую точку бифуркации

```python
def func(y, t, params):
    r, K = params
    eq =   r* y *(1 - y/K) - y**2 / (1 + y**2)
    return eq

tt = np.linspace(0, 100)
y0 = 8

fig = plt.figure(figsize = (10, 4))

r_pars = [0.5, 0.4, 0.3, 0.2, 0.1]
for r in r_pars:
    rK = [r, 15]
    Yint = slv.odeint(func, y0, tt, args=(rK,))
    plt.plot(tt, Yint, label = f'r = {r}')
    
plt.annotate('Вымирание при r меньше значения точки бифуркации',
             (17, 4),xytext=(30, 9),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)

plt.title('Y(t), точка бифуркации примерно r =  0.27')
plt.legend(bbox_to_anchor=(1, 1))
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

### П. 6

Кинетики y(t) при маленьком y и r, проходящем через правую точку бифуркации


```python
def func(y, t, params):
    r, K = params
    eq =   r* y *(1 - y/K) - y**2 / (1 + y**2)
    return eq

tt = np.linspace(0, 100)
y0 = 1

fig = plt.figure(figsize = (10, 4))

r_pars = [0.3, 0.4, 0.5, 0.536, 0.6, 1]
for r in r_pars:
    rK = [r, 15]
    Yint = slv.odeint(func, y0, tt, args=(rK,))
    plt.plot(tt, Yint, label = f'r = {r}')
    
plt.annotate('Резкий рост при r больше значения точки бифуркации',
             (17, 8),xytext=(25, 5),
             arrowprops= {'width': 0.1, 'headwidth': 5},
             fontsize= 13)

plt.title('Y(t), точка бифуркации примерно r = 0.536')
plt.legend(bbox_to_anchor=(1, 1))
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
