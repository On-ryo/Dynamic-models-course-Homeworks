### 1.

$$
\left\{\begin{array}{l}
\frac{d x}{d t}=3 x \\
\frac{d y}{d t}=-2 x-2 y
\end{array}\right.
$$

```python
P = lambda x,y: +3*x
Q = lambda x,y: -2*x - 2*y
fun0 = lambda p,t: [ P(p[0],p[1]), Q(p[0],p[1]) ]
```

```python
# фазовые траектории

# задать время
tt = np.linspace(0, 1, 300)

x0s = np.linspace(-1, 1, 5)
y0s  = np.linspace(-1, 1, 5)

plt.figure(figsize = (7,5))
for y0 in y0s:
    for x0 in x0s:
        xy0 = [x0, y0]

        zz = slv.odeint(fun0, xy0, tt)
        plt.plot(zz[:,0], zz[:,1])
        plt.plot([xy0[0]], [xy0[1]], 'o', mec='k', ms=4)

plt.title('фазовые траектории')
plt.axhline(0, color='b', linestyle='--', lw = 1)
#plt.axvline(0, color='b', linestyle='--', lw = 1)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.ylabel('y')
```

```python
# кинетики

xy0s = [(-1, 0), (0, -1), (1,1), (0,0)]

tt = np.linspace(0, 10, 300)
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))

i = 0

for ax in axes.flat:

    xy0 =  xy0s[i]
    zz = slv.odeint(fun0, xy0, tt)
    ax.plot(tt, zz[:,0], label = 'x')
    ax.plot(tt, zz[:,1], label = 'y')
    ax.axhline(y=0, color='b', linestyle='--', lw = 1)
    ax.set_title(f'x0 = {xy0[0]}, y0 ={xy0[1]}')
    ax.set_xlabel('t')
    ax.set_xlim(-0.1, 5)
    ax.set_ylim(-10, 10)
    ax.legend()


    i += 1




```

<font color='red'>+</font> 

```python
# векторное поле

xx = np.linspace(-5,5,20)
yy = np.linspace(-4,4,20)
Xm,Ym = np.meshgrid(xx,yy)

plt.quiver(Xm,Ym, P(Xm,Ym), Q(Xm,Ym), scale_units='xy', angles='xy')

#  собственные векторы
mtx = np.array([[3, 0],[-2,-2]])
vals, vecs = np.linalg.eig(mtx)
plt.plot([vecs[0,0]*-5, vecs[0,0]*5] , [vecs[1,0]*-5, vecs[1,0]*5], 
             '--r', label='$\lambda_1$ = %.3f' % vals[0])
plt.plot([vecs[0,1]*-5, vecs[0,1]*5] , [vecs[1,1]*-5, vecs[1,1]*5], 
             '--b', label='$\lambda_2$ = %.3f' % vals[1])

# decorations
plt.axhline(0, color='k', linestyle='--', lw = 1)

plt.legend()
plt.xlim(-5,5)
plt.ylim(-4,4)
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
def display2Deq(eq1, eq2):
    s = \
    r'''$$
    \left\{
    \begin{aligned}
        %s \\
        %s
    \end{aligned}
    \right.
    $$''' % (sym.latex(eq1).replace('=','&='), sym.latex(eq2).replace('=','&='))
    display(Latex(s))
```

```python
# задать переменные

p, t, x0, y0 = sym.symbols("p, t, x0, y0")
```

```python
x = sym.Function('x')(t)
y = sym.Function('y')(t)
```

```python
x_rate = -(2 + p)*x - 5*y
y_rate = 2*x + 2*y
```

```python
eq1 = sym.Eq(x.diff(t), x_rate)
eq2 = sym.Eq(y.diff(t), y_rate)
```

```python
display2Deq(eq1,eq2)
```

```python
solved_eqs = sym.dsolve([eq1, eq2],
          ics={x.subs(t,0): x0,
               y.subs(t,0): y0})
```

```python
display(solved_eqs[0])
display(solved_eqs[1])
```

```python
xlam = sym.lambdify([t,p,x0,y0], solved_eqs[0].rhs)
ylam = sym.lambdify([t,p,x0,y0], solved_eqs[1].rhs)
```

```python
#  кинетики для разных p 

for pv in [3, 4, 10]:
    tt = np.linspace(0, 10, 1000)
    xx = xlam(tt, pv, -1, -1)
    yy = ylam(tt, pv, -1,-1)

    plt.plot(tt, xx, '-', label = f'x, p = {pv}' )
    plt.plot(tt, yy, '--', label = f'y, p = {pv}')


plt.xlim(0, 10)
plt.ylim(-10,20)
_ = plt.xlabel('t')
_ = plt.ylabel('x or y')
_ = plt.legend(bbox_to_anchor=(1.4, 1))
```

```python
# фазовые траектории для разных p


for pv in [3, 4, 10]:
    tt = np.linspace(0, 10, 100)
    xx = xlam(tt, pv, 3, 3)
    yy = ylam(tt, pv, 3, 3)

    plt.plot(xx, yy, label = f'p = {pv}' )
    
plt.xlim(-15,5)
plt.ylim(-1,20)
plt.title('фазовые траектории для разных p')
_ = plt.legend()
_ = plt.xlabel('x')
_ = plt.ylabel('y')
```

#### Поиск стационарных точек

```python
display2Deq(
    sym.Eq(0, x_rate), 
    sym.Eq(0, y_rate) )
```

```python
# стационарная точка: 
sym.nonlinsolve([x_rate, y_rate], [x,y])
```

```python
# матрица Якоби:

J = sym.Matrix([x_rate, y_rate]).jacobian(sym.Matrix([x,y]))
J
```

```python
# собственные числа и векторы: 
vecs = J.eigenvects()
vecs
```

```python
# собственные числа и их кратность 
print(J.eigenvects()[0][0])
print(J.eigenvects()[0][1])

print(J.eigenvects()[1][0])
print(J.eigenvects()[1][1])
```

```python
# собственный вектор 1
J.eigenvects()[0][2][0]
```

```python
# собственный вектор 2
J.eigenvects()[1][2][0]
```

```python
# зависимость вещественной части собственных чисел от параметра

lam1 = sym.lambdify(p, sym.re(J.eigenvects()[0][0]) )
lam2 = sym.lambdify(p, sym.re(J.eigenvects()[1][0]) )
pp = np.linspace(-5,5,100)

plt.figure(figsize = (15, 5))
plt.subplots(1, 3, sharey = True)

plt.subplot(131)
plt.plot(pp, lam1(pp), label = '$\lambda_1$')
plt.axhline(0, color='k')
plt.grid()
_ = plt.xlabel('p')
_ = plt.ylabel('$\lambda$')
_ = plt.legend()

plt.subplot(132)
plt.plot(pp, lam2(pp), color='orange', label = '$\lambda_2$')
plt.axhline(0, color='k')
plt.grid()
_ = plt.xlabel('p')
_ = plt.ylabel('$\lambda$')
_ = plt.legend()

plt.subplot(133)
plt.plot(pp, lam1(pp), label = '$\lambda_1$')
plt.plot(pp, lam2(pp), label = '$\lambda_2$')
plt.axvline(x=3, color='blue', lw= 0.5, label = 'p=3')
plt.grid()
plt.axhline(0, color='k')
_ = plt.xlabel('p')
_ = plt.ylabel('$\lambda$')
_ = plt.legend()



```

при $p < 0$ оба собственных числа больше нуля - точка неустойчива + совпадают значения (неустойчивый фокус)  
при $0 <p < примерно 2 $ - собственные числа имеют однаковый знак - точка устойчива  + совпадают значения (устойчивый фокус)  
при $ примерно 2 <p < 3 $ -  собственные числа имеют однаковый знак - точка устойчива  + разные значения (устойчивый узел)  
при $ 3 < p  $ собств. числа имеют разные знаки  (точка типа седло).
